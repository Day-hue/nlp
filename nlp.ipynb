```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download necessary resources
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

1. Text Cleaning (Preprocessing with Regex)
  * Convert text to lowercase (if case sensitivity is not required).
  * Remove special characters, punctuation, extra spaces, etc.
 * Remove refrences
 * Handle contractions (e.g., "don't" → "do not").
 * Normalize text (if necessary, such as removing accents).

```python
#convert text to lowercase
text = text.lower()
print(text)
```

```python
import re

def remove_references(text):
    text = re.sub(r"\[\d+\]", "", text)  # Remove single references like [17]
    text = re.sub(r"\[\d+(,\d+)*\]", "", text)  # Remove multiple references like [145,146]
    text = re.sub(r"\s+", " ", text).strip()  # Remove extra spaces
    return text
cleaned_text = remove_references(text)
print(cleaned_text)
```

```python
#removing special characters
# Removes only brackets
cleaned_text = re.sub(r"[\[\]()]","", cleaned_text)
print(cleaned_text)
```

```python
# replace dashes with spaces (to keep words separate)
cleaned_text = re.sub(r"[-–—]", " ", cleaned_text)
print(cleaned_text)
```

```python
# Removes dollar signs
cleaned_text = re.sub(r"\$", "", cleaned_text)
print(cleaned_text)
```

```python
# Removes all punctuation except full stops (.) and percentage signs (%)
cleaned_text = re.sub(r"[^\w\s%.]", "", cleaned_text)
print(cleaned_text)
```

# 2. Tokenization

Split the text into words (word tokenization) or sentences (sentence tokenization).
The choice between sentence tokenization and word tokenization depends on the specific NLP task you're working on. Here's when to use each:

 ## 1. Sentence Tokenization (Splitting text into sentences)

### Use when:
 * You need to analyze text at the sentence level (e.g., sentiment analysis, summarization, translation).
 * Context matters, and breaking text into individual words would lose meaning.
 * You're working with long documents and want to process one sentence at a time.

### Avoid when:
 * You're working on tasks that require analyzing words independently (e.g., keyword extraction, named entity recognition).

### Example:
 * Input:

      "I love NLP. It's amazing!"

 * Output:
      
      "I love NLP."

      "It's amazing!"

   ## 2. Word Tokenization (Splitting text into words)

### Use when:
 * You're working with bag-of-words, TF-IDF, Word2Vec, or any model that requires word-level features.
 * You need to remove stopwords, perform stemming/lemmatization, or extract meaningful terms.
 * Your focus is on individual words rather than sentence structure.

### Avoid when:
 * You need sentence structure for meaning preservation.

### Example:

 * Input:

     "I love NLP."

 * Output:
      
      "I"
      
      "love"
      
      "NLP"

### Which One to Use?
 * If you're working on word-based analysis (e.g., text classification, topic modeling, embeddings) → Word Tokenization
 * If you're working on sentence-level tasks (e.g., summarization, machine translation, sentiment at the sentence level) → Sentence Tokenization
 * Sometimes, you may use both (e.g., first sentence tokenization, then word tokenization on each sentence).

```python
#sentense tokenization
sentences = sent_tokenize(cleaned_text)
print(sentences)
```

```python
#word tokenization
words = word_tokenize(cleaned_text)
print(words)
```

# 3. Stopwords Removal
Remove common words that do not add significant meaning (e.g., "the," "is," "and").
Optional: Keep stopwords if they hold meaning in context (e.g., sentiment analysis).

```python
# Remove stopwords
stop_words = set(stopwords.words("english"))
filtered_words = [word for word in words if word.lower() not in stop_words]

print("Original Words:", words)
print("Filtered Words:", filtered_words)
```

# 4. Lemmatization or Stemming
Lemmatization:
 * Convert words to their base form using vocabulary-based techniques (better for linguistic accuracy).

Stemming:
 * Reduce words to their root form using heuristic rules (faster but less accurate).
Use either stemming or lemmatization, not both.

```python
nltk.download("omw-1.4")
# Initialize Lemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatize each word in the filtered word list
# 'pos="v"' specifies that the words should be treated as verbs during lemmatization
lemmatized_words = [lemmatizer.lemmatize(word, pos="v") for word in filtered_words]  # 'v' for verb
# Print the lemmatized words
print(lemmatized_words)
```

```python
# Initialize the Porter Stemmer
stemmer = PorterStemmer()
# Apply stemming to each word in the filtered word list
stemmed_words = [stemmer.stem(word) for word in filtered_words]
# Print the stemmed words
print(stemmed_words)
```

# Bag of Words (BOW)
## What is Bag of Words (BoW)?
Bag of Words (BoW) is a text representation technique used in Natural Language Processing (NLP) and Machine Learning.

It converts text into a numerical format that can be used by algorithms.

## Key Idea:

A document is represented as a "bag" of words (i.e., a collection of words without considering order or grammar).

The focus is on word frequency (i.e., how many times a word appears in the text).

BOW creates a word-frequency matrix where each row represents a document, and each column represents a word.

```python
from sklearn.feature_extraction.text import CountVectorizer


vectorizer = CountVectorizer()
X = vectorizer.fit_transform(words)

print(vectorizer.get_feature_names_out())  # Show vocabulary
print(X.toarray())  # Show word frequencies
```

# **TF-IDF (Term Frequency - Inverse Document Frequency) **

TF-IDF is a powerful text representation technique used in **Natural Language Processing (NLP)** to measure the importance of words in a document relative to a collection of documents (**corpus**).

---

## **📌 Why Use TF-IDF?**
Unlike **Bag of Words (BoW)**, which treats all words equally, **TF-IDF assigns more weight** to important words and reduces the influence of common words like *"the," "is,"* and *"and."*

---

## **1️⃣ How TF-IDF Works**
TF-IDF is a combination of two metrics:

### **1. Term Frequency (TF)**
Measures how frequently a word appears in a document.

$$
TF = \frac{\text{Number of times a word appears in a document}}{\text{Total words in that document}}
$$

#### **Example:**
For the sentence:  
`"Python is great and Python is powerful"`

The TF of `"Python"` is:

$$
TF("Python") = \frac{2}{6} = 0.33
$$

---

### **2. Inverse Document Frequency (IDF)**
Measures how important a word is across multiple documents.

$$
IDF = \log\left(\frac{\text{Total number of documents}}{\text{Number of documents containing the word}}\right)
$$

- **Rare words** get **higher IDF scores** (important words).  
- **Common words** get **lower IDF scores** (*stopwords like "is" and "the"*).

#### **Example:**
If the word `"Python"` appears in **1 out of 10** documents:

$$
IDF("Python") = \log\left(\frac{10}{1}\right) = 1.0
$$

---

### **3. TF-IDF Score Calculation**
The **TF-IDF score** is the product of **TF** and **IDF**:

$$
TF-IDF = TF \times IDF
$$

#### **Example (for `"Python"`):**

$$
TF-IDF("Python") = 0.33 \times 1.0 = 0.33
$$

---

✅ **TF-IDF is widely used in:**  
- **Search engines** (ranking documents based on relevance).  
- **Text classification** (spam detection, topic modeling).  
- **Keyword extraction** (finding important words in text).  
- **Recommendation systems** (suggesting articles, books, etc.).

# n-Grams (Unigrams, Bigrams, Trigrams) in NLP
What are n-Grams?
 * An n-Gram is a contiguous sequence of n words from a given text. It is a fundamental concept in Natural Language Processing (NLP) used for text analysis, speech recognition, and machine learning models.

 * Unigram (1-Gram) → Single word
 * Bigram (2-Gram) → Two consecutive words
 * Trigram (3-Gram) → Three consecutive words
 * n-Gram (n-Gram) → Sequence of n words

## Why Use n-Grams?
 * Captures Word Relationships: Unlike Bag of Words (BoW), which ignores word order, n-Grams preserve local context.
 * Improves Language Models: Helps predict the next word in a sentence (e.g., autocomplete).
 * Useful for Feature Engineering: Used in text classification, machine translation, and sentiment analysis.
 * Detects Common Phrases: Useful for identifying frequent word combinations (e.g., "New York" or "artificial intelligence").


# Types of n-Grams
## Unigram (n = 1)
 * A unigram is a single word.
 * Captures word frequency but loses context.
 * Often used in BoW and TF-IDF models.

## Bigram (n = 2)
 * A bigram is a sequence of two consecutive words.
 * Captures basic word relationships.
 * Useful for detecting common phrases.

## Trigram (n = 3)
 * A trigram consists of three consecutive words.
 * More context than bigrams but needs more data.
 * Used in chatbots, predictive text, and speech recognition.

## n-Gram (General Case)
 * For an n-Gram, words are grouped in sequences of n.
 * Higher n captures more context but increases sparsity.

We can generate n-Grams using NLTK and scikit-learn.

Method 1: Using NLTK

```python
from nltk import ngrams
# Generate n-Grams
unigrams = list(ngrams(words, 1))
bigrams = list(ngrams(words, 2))
trigrams = list(ngrams(words, 3))

print("Unigrams:", unigrams)
print("Bigrams:", bigrams)
print("Trigrams:", trigrams)
```

Method 2: Using scikit-learn (CountVectorizer)

```python
from sklearn.feature_extraction.text import CountVectorizer


# Unigram representation
vectorizer_uni = CountVectorizer(ngram_range=(1, 1))
X_uni = vectorizer_uni.fit_transform(sentences)
print("Unigrams:", vectorizer_uni.get_feature_names_out())

# Bigram representation
vectorizer_bi = CountVectorizer(ngram_range=(2, 2))
X_bi = vectorizer_bi.fit_transform(sentences)
print("Bigrams:", vectorizer_bi.get_feature_names_out())

# Trigram representation
vectorizer_tri = CountVectorizer(ngram_range=(3, 3))
X_tri = vectorizer_tri.fit_transform(sentences)
print("Trigrams:", vectorizer_tri.get_feature_names_out())
```
